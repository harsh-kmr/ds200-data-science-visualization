```python
# data link
# https://www.data.gov.in/resource/area-population-and-income-g-20-countries-upto-2013#api

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_file_url = "https://raw.githubusercontent.com/harsh-kmr/ds200-data-science-visualization/refs/heads/main/Area%2C%20Population%20And%20Income%20Of%20G-20%20Countries%20upto%202013.csv"
data = pd.read_csv(csv_file_url)
data.head()
```

```python
data.columns
```

```python
# filter top 5 country by gdp 2013
data_filtered = data.sort_values(
    by="GDP Billions (US$) - 2013", ascending=False
).head(7)

# Remove World and EU
data_filtered = data[(data['Country Name'] != 'World') &
                     (data['Country Name'] != 'European Union')]

# Plot
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6), num=1, dpi=300)

scatter = sns.scatterplot(
    data=data_filtered,
    x="Population (Millions) - 2013",
    y="GDP Billions (US$) - 2013",
    hue="Country Name",
    alpha=0.7,
    edgecolor="black"
)

plt.title("GDP vs Population (Top 5 Countries by GDP, 2013)", fontsize=14, fontweight="bold")
plt.xlabel("Population (Millions, 2013)", fontsize=12)
plt.ylabel("GDP (Billions US$, 2013)", fontsize=12)

plt.legend(title="Country", loc="upper right", fontsize=9)

plt.grid(True, linestyle="--", alpha=0.6)
plt.tick_params(axis="both", which="major", labelsize=10)
plt.minorticks_on()
plt.tick_params(axis="both", which="minor", length=4, color="gray")

plt.tight_layout()
plt.show()
```

```python
# Create figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6), dpi=300)

# Remove World and EU
data_filtered = data[(data['Country Name'] != 'World') &
                     (data['Country Name'] != 'European Union')]

# Population boxplot
sns.boxplot(
    data=data_filtered,
    y="Population (Millions) - 2013",
    ax=axes[0],
    color="skyblue"
)
axes[0].set_title("Population (Millions, 2013)", fontsize=12, fontweight="bold")
axes[0].set_ylabel("Population (Millions)")
axes[0].set_xlabel("Countries")
axes[0].grid(True, linestyle="--", alpha=0.6)

# GDP boxplot
sns.boxplot(
    data=data_filtered,
    y="GDP Billions (US$) - 2013",
    ax=axes[1],
    color="lightgreen"
)
axes[1].set_title("GDP (Billions US$, 2013)", fontsize=12, fontweight="bold")
axes[1].set_ylabel("GDP (Billions US$)")
axes[1].set_xlabel("Countries")
axes[1].grid(True, linestyle="--", alpha=0.6)

# DP per capita boxplot
sns.boxplot(
    data=data_filtered,
    y="Domestic Product Per Capita Income (US$) - 2013",
    ax=axes[2],
    color="salmon"
)
axes[2].set_title("DP Per Capita (US$, 2013)", fontsize=12, fontweight="bold")
axes[2].set_ylabel("DP Per Capita (US$)")
axes[2].set_xlabel("Countries")
axes[2].grid(True, linestyle="--", alpha=0.6)

for ax in axes:
    for spine in ax.spines.values():
        spine.set_visible(True)

plt.suptitle("Box Plots of Population, GDP, and GDP per Capita (2013)", fontsize=14, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Filter data for selected countries
countries = ["India", "USA", "China", "Japan"]
data_filtered = data[data["Country Name"].isin(countries)]

gdp_cols = [
    "GDP Billions (US$) - 2010",
    "GDP Billions (US$) - 2011",
    "GDP Billions (US$) - 2012",
    "GDP Billions (US$) - 2013"
]
data_melted = data_filtered.melt(
    id_vars=["Country Name"],
    value_vars=gdp_cols,
    var_name="Year",
    value_name="GDP (Billions US$)"
)

data_melted["Year"] = data_melted["Year"].str.extract(r"(\d{4})")

# Plot
fig, ax = plt.subplots(figsize=(10, 6), dpi=300)

sns.lineplot(
    data=data_melted,
    x="Year",
    y="GDP (Billions US$)",
    hue="Country Name",
    marker="o",
    ax=ax
)

ax.set_title("GDP over Time (2010–2013)", fontsize=14, fontweight="bold")
ax.set_xlabel("Year", fontsize=12)
ax.set_ylabel("GDP (Billions US$)", fontsize=12)

ax.tick_params(axis="x", which="both", labelsize=10)
ax.tick_params(axis="y", which="both", labelsize=10)
ax.grid(True, linestyle="--", alpha=0.6)
ax.legend(title="Country", fontsize=10, title_fontsize=11, loc = [0.87,0.70])  # Legend

for spine in ax.spines.values():
    spine.set_visible(True)
    spine.set_linewidth(1)
    spine.set_color("black")

plt.tight_layout()
plt.show()
```
